```python
from io import StringIO
cols = ['extraction','drug','meddra_term']
data = data[cols]
data.columns = ['extraction','drug','meddra_term']
data=data.dropna()
data.head(50)

cols1 = ['extraction','drug','meddra_term']
data1 = data1[cols1]
data1.columns = ['extraction','drug','meddra_term']
data1=data1.dropna()
data1.head(50)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
nb = Pipeline([('vect', CountVectorizer()),
               ('tfidf', TfidfTransformer()),
               ('clf', MultinomialNB()),
              ])
```

```python
x=data.iloc[:,0]
x
x1=data1.iloc[:,0]
x1
```

```python
y=data.iloc[:,2]
y
y1=data1.iloc[:,2]
y1
```

```python
nb.fit(x,y)
```

```python
from sklearn.metrics import classification_report
y_pred = nb.predict(x1)
from sklearn.metrics import accuracy_score, confusion_matrix

print('accuracy %s' % accuracy_score(y_pred,y1))
#print(classification_report(y1, y_pred))
```
